```python
import yaml

# Define the dataset structure and classes
data = {
    "path": "/content/drive/MyDrive/dataset",  # Path to your dataset
    "train": "images/train",
    "val": "images/val",
    "test": "images/test",
    "names": [
        "2_taka_note", "5_taka_note", "10_taka_note", "20_taka_note", "50_taka_note",
        "100_taka_note", "200_taka_note", "500_taka_note", "1000_taka_note",
        "2_taka_coin", "5_taka_coin"
    ]
}

# Save to a .yaml file
yaml_path = "/content/drive/MyDrive/dataset/data.yaml"
with open(yaml_path, 'w') as f:
    yaml.dump(data, f, sort_keys=False)

print(f"data.yaml written to {yaml_path}")
```

```python
!pip install ultralytics

from ultralytics import YOLO

# Load a pre-trained YOLO model or use your own best.pt
model = YOLO("yolov8n.pt")  # Or use "best.pt" if you have it

# Train with your updated data.yaml file
model.train(
    data="/content/drive/MyDrive/dataset/data.yaml",  # Path to your updated data.yaml
    epochs=80,         # Number of epochs
    batch=16,          # Batch size (adjust as needed)
    imgsz=640,         # Image size
    device="cpu",      # Use "0" for GPU if available
    project="/content/drive/MyDrive/runscoins",  # Save results in a separate folder (runscoins)
    name="train_with_coins"  # Name of this run
)
```

```python
# Evaluate the model on the test dataset
metrics = model.val(data="/content/drive/MyDrive/dataset/data.yaml", split="test")

# Print out evaluation metrics
print("Precision:", metrics.box.mp)
print("Recall:", metrics.box.mr)
print("mAP50:", metrics.box.map50)
print("mAP50-95:", metrics.box.map)
```
